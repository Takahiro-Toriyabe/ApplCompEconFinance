```python
# Exercise 1.2 (a)
A = np.matrix([[0, -1, 2], [-2, -1, 4], [2, 7, 3]])
B = np.matrix([[-7, 1, 1], [7, -3, -2], [3, 5, 0]])
y = np.array([3, -1, 2])
C1 = A @ B
x1 = np.linalg.solve(C1, y)
print('C1 =')
print(C1)

print('x1 =')
print(x1)
```

```python
# Exercise 1.2 (b)
C2 = np.multiply(A, B)
x2 = np.linalg.solve(C2, y)
print('C =')
print(C2)

print('x =')
print(x2)
```

```python
%matplotlib inline

# Exercise 1.3
np.random.seed(seed=1120002)

t = np.array(list(range(1960, 2002)))
one = np.array([1]*len(t))
e = np.random.normal(loc=0, scale=0.2, size=len(t))
y = 5 + 0.05 * t + e

x = np.array([one.T, t.T])
b = np.linalg.solve(x @ x.T, x @ y)
yhat = x.T @ b

fig, ax = plt.subplots()
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.grid()
ax.plot(t, y, color='black', label='actual')
ax.plot(t, yhat, color='red', label='fitted')
ax.legend(loc=0) 
fig.tight_layout()
plt.show()
```

**Exercise 1.4 (a)**
Without the government support payment,
$$
E[p] = 3 - 2 \left(0.5 + 0.5 E[p] \right) E[\tilde{y}]
$$
Since $E[\tilde{y}] = 0.5 * 0.7 + 0.5 * 1.3 = 1$, the expected equilibrium price is $E[p] = 1$. Similarly, we observe
$$
Var(p) = 4\left(0.5 + 0.5 E[p] \right)^2 Var(\tilde{y}) = 4 Var(\tilde{y}) = 0.36
$$

**(b)** Under the government support payment, we observe
$$
E[p] = 2 - E[\max(p, 1)]; \quad Var(p) = 0.09(1 + E[\max(p, 1)])^2
$$


```python
# (b) (continued)
y = np.array([0.7, 1.3])
w = np.array([0.5, 0.5])
a = 1
aold = 1000
while abs(a - aold) > 1.0e-8:
    aold = a
    p = 3 - 2 * a * y
    f = np.dot(w, np.vectorize(lambda x: max(x, 1))(p))
    a = 0.5 + 0.5 * f

print('Equilibrium acreage = {0:.2f}'.format(a))
print('Expected market price = {0:.2f}'.format(np.dot(w, p)))
print('Expected producers price = {0:.2f}'.format(f))
```

```python
# Exercise 1.4 (c)
s = (f - np.dot(w, p)) / a
print('Expected vovernment subsidy per planted acre = {0:.2f}'.format(s))
```
